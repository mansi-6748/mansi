```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.show()
```

```python
sns.lmplot(x="area", y="price", data=df)
plt.show()
```

```python
# Create linear regression object
reg = linear_model.LinearRegression() # hotcoded
reg.fit(df[['area']], df['price'])
```

```python
reg.predict([[3300]])
```

```python
joblib.dump(reg, 'house_price_model.pkl')
lm = joblib.load('house_price_model.pkl')
```

```python
lm.predict([[3300]])
```
